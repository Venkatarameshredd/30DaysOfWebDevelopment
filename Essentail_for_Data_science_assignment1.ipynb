```python
#Take datasets in different formats (CSV, Excel, JSON, etc.) and perform data cleaning by handling empty cells, wrong formats,
#inconsistent data, and duplicates. Demonstrate missing data handling using functions like isnull(), notnull(), dropna(), fillna(), replace(), and interpolate().
# Dataset1
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/heart_disease.csv")
print(df)
#data cleaning
#1.Check for Missing Values
print(df.isnull())
#2.detecting  existing non missing values
print(df.notnull())

#2Drop Columns with Missing Values
print(df.dropna(axis=1))

#3.Check for Duplicate Rows
print(df.duplicated())

#4.Count Missing Values Per Column
print(df.isnull().sum())

#5.Fill Missing Values with a Constant
print(df.fillna(5))

#6replace missing values
print(df.replace(to_replace=np.nan))

#7.Fill Missing Values by Interpolation

print(df.interpolate())
```
